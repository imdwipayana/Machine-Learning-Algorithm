```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bunga_iris = pd.read_csv('https://raw.githubusercontent.com/imdwipayana/DATA/main/Bunga%20Iris/bunga_iris.csv')
bunga_target = pd.read_csv('https://raw.githubusercontent.com/imdwipayana/DATA/main/Bunga%20Iris/bunga_iris_target.csv')
```

```python
bunga_1 = bunga_iris.loc[bunga_target['target']==0]
bunga_2 = bunga_iris.loc[bunga_target['target']==1]
bunga_3 = bunga_iris.loc[bunga_target['target']==2]
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.scatter(bunga_1['sepal length (cm)'],bunga_1['sepal width (cm)'])
plt.scatter(bunga_2['sepal length (cm)'], bunga_2['sepal width (cm)'])
plt.scatter(bunga_3['sepal length (cm)'], bunga_3['sepal width (cm)'])

plt.show()
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')

plt.scatter(bunga_1['petal length (cm)'], bunga_1['petal width (cm)'])
plt.scatter(bunga_2['petal length (cm)'], bunga_2['petal width (cm)'])
plt.scatter(bunga_3['petal length (cm)'], bunga_3['petal width (cm)'])

plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(bunga_iris,bunga_target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```

```python
# Tune Parameter
```

```python
# Regularization(C)
model_C = SVC(C=1)
model_C.fit(X_train,y_train)
model_C.score(X_test,y_test)
```

```python
model_C = SVC(C=10)
model_C.fit(X_train,y_train)
model_C.score(X_test,y_test)
```

```python
model_C = SVC(C=4)
model_C.fit(X_train,y_train)
model_C.score(X_test,y_test)
```

```python
# Gamma
model_G = SVC(gamma=10)
model_G.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
# Kernel
model_kernel = SVC(kernel='linear')
model_kernel.fit(X_train,y_train)
model_kernel.score(X_test,y_test)
```
