```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
```

# K-Nearest Neighbors dengan menggunakan data bunga Iris
#### Terdapat tiga jenis bunga Iris yaitu Sentosa, Versicolor dan Virginica. Ketiganya memiliki bentuk dan corak warna yang hampir serupa. Untuk itu akan digunakan algoritma KNN untuk mengklasifikasikan bunga Iris apakah termasuk jenis Sentosa, Versicolor atau Virginica dengan menggunakan data panjang sepal, lebar sepal, panjang petal dan lebar petal bunganya.


```python
bunga_iris = pd.read_csv('https://raw.githubusercontent.com/imdwipayana/DATA/main/Bunga%20Iris/bunga_iris.csv')
bunga_target = pd.read_csv('https://raw.githubusercontent.com/imdwipayana/DATA/main/Bunga%20Iris/bunga_iris_target.csv')
```

```python
bunga_iris.head()
```

```python
bunga_target.head()
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
bunga_1 = bunga_iris.loc[bunga_target['target']==0]
bunga_2 = bunga_iris.loc[bunga_target['target']==1]
bunga_3 = bunga_iris.loc[bunga_target['target']==2]
plt.scatter(bunga_1['sepal length (cm)'],bunga_1['sepal width (cm)'], color ='green', marker='+')
plt.scatter(bunga_2['sepal length (cm)'],bunga_2['sepal width (cm)'], color = 'red', marker='.')
plt.scatter(bunga_3['sepal length (cm)'],bunga_3['sepal width (cm)'], color = 'yellow', marker='*')
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
bunga_1 = bunga_iris.loc[bunga_target['target']==0]
bunga_2 = bunga_iris.loc[bunga_target['target']==1]
bunga_3 = bunga_iris.loc[bunga_target['target']==2]
plt.scatter(bunga_1['petal length (cm)'],bunga_1['petal width (cm)'], color ='green', marker='+')
plt.scatter(bunga_2['petal length (cm)'],bunga_2['petal width (cm)'], color = 'red', marker='.')
plt.scatter(bunga_3['petal length (cm)'],bunga_3['petal width (cm)'], color = 'yellow', marker='*')
```

```python
X_train, X_test, y_train, y_test = train_test_split(bunga_iris,bunga_target,test_size=0.2, random_state=1)
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
knn.predict([[1,2,3,4]])
```
